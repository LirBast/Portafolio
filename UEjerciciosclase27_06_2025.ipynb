```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("spscientist/students-performance-in-exams")

print("Path to dataset files:", path)
```

```python
import os
import pandas as pd

os.listdir("/kaggle/input/students-performance-in-exams")
path2=f"/kaggle/input/students-performance-in-exams/StudentsPerformance.csv"

df = pd.read_csv(path2)
print(df)
print("------------------------------------------------------------")
print(df.head())
print("------------------------------------------------------------")
print(df.info())
print("------------------------------------------------------------")
print(df.describe())
print("------------------------------------------------------------")
print(df.isnull().sum())  ### los metodos llevan parentesis
print("------------------------------------------------------------")

q1=df["math score"].quantile(0.25)
print(q1)
print("------------------------------------------------------------")
q3=df["math score"].quantile(0.75)
print(q3)
print("------------------------------------------------------------")
iqr=q3-q1
print(f"el valor del IQR es {iqr}")
limt_inf=q1-1.5*iqr
limt_sup=q3+1.5*iqr
print(f"el valor del limite inferior es {limt_inf}")
print(f"el valor del limite superior es {limt_sup}")

print("------------------------------------------------------------")
outliers = df[(df["math score"] < limt_inf) | (df["math score"] > limt_sup)]
print(outliers)

df_sin_outliers = df[(df["math score"] >= limt_inf) & (df["math score"] <= limt_sup)]
print(df_sin_outliers)
```
